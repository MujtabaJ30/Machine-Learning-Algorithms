```python
df=pd.read_csv("bhp.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
upper_limit=df.price_per_sqft.mean()+ 3*df.price_per_sqft.std()
lower_limit=df.price_per_sqft.mean()- 3*df.price_per_sqft.std()
```

```python
df.price_per_sqft.mean()
```

```python
df.shape
```

```python
lower_limit
```

```python
upper_limit
```

```python
df2=df[(df.price_per_sqft<=upper_limit)&(df.price_per_sqft>=lower_limit)]
```

```python
df2.shape
```

```python
df2['zscore']=(df2.price_per_sqft-df2.price_per_sqft.mean())/df2.price_per_sqft.std()
```

```python
df2.head(50)
```

```python
df2 = df2[~((df2['zscore'] <= -3) | (df2['zscore'] >= 3))]
```

```python
df2.shape
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.hist(df2.price_per_sqft, bins=20,rwidth=0.8)
plt.show()
```

```python
from scipy.stats import norm
import numpy as np

plt.hist(df2.price_per_sqft, bins=20, rwidth=0.8, density=True)
plt.xlabel('Height (inches)')
plt.ylabel('Count')

rng = np.arange(-5000, df2.price_per_sqft.max(), 100)
plt.plot(rng, norm.pdf(rng,df2.price_per_sqft.mean(),df2.price_per_sqft.std()))
```
