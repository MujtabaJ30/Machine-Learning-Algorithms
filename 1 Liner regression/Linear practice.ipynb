```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("canada_income.csv")
```

```python
df.head()
```

```python
df.rename(columns={'per capita income (US$)':'income'}, inplace=True)
```

```python
%matplotlib inline
plt.xlabel('Per Capita Income (US$)')
plt.ylabel("Year")
plt.scatter(df.income,df.year)
```

```python
reg=LinearRegression()
reg.fit(df[['year']],df.income)
```

```python
x=2021
reg.predict([[x]])
```

```python
d=pd.DataFrame({'year':["2017","2018","2019","2020","2021","2022","2023","2024","2025","2026","2027"]})
```

```python
d.to_csv("years.csv")
```

```python
p=reg.predict(d)
```

```python
d['income']=p
```

```python
d.head(2)
```

```python
%matplotlib inline 
plt.xlabel('Income')
plt.ylabel('year')
plt.scatter(df['income'],df['year'], label="Training")
plt.plot(reg.predict(df.income),d['year'],label="Predicted",color='red',linestyle='dashed')

plt.legend()
plt.show()
```
