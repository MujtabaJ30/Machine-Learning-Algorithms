```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score , classification_report
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("titanic.csv")
```

```python
df.head(2)
```

```python
df.drop(columns=['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], inplace=True)
```

```python
df.head(2)
```

```python
print(df.isnull().sum())
```

```python
df.Age = df.Age.fillna(df.Age.median())
```

```python
df.tail()
```

```python
target=df.Survived
input=df.drop(columns=['Survived'])
```

```python
le_Sex=LabelEncoder()

input['Sex']=le_Sex.fit_transform(df.Sex)
```

```python
input.head(2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(input,target,test_size=0.25 )
```

```python
X_train
```

```python
model.fit(X_train,Y_train)
```
